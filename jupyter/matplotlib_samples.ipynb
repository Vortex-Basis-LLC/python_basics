# matplotlib

- [Matplotlib Examples](https://matplotlib.org/stable/gallery/index.html)

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Histogram showing counts in buckets of point totals for points scored each game.
points_for_games = [10, 12, 5, 12, 3, 1, 2, 8, 7, 19, 22, 13, 15, 12, 16]

plt.figure(figsize=(7, 3))
plt.title("Point Totals")
plt.xlabel("Points")
plt.ylabel("Count")
plt.hist(points_for_games, rwidth=0.9, color="blue", bins=range(0, 25, 6))
plt.show()
```
