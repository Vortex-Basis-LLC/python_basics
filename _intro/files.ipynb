```python
# A few ways to list files
import os
from os import path
import glob

print("os.listdir")
for name in os.listdir("."):
    print("file:" if path.isfile(name) else "dir:", name)

print("")

print("os.walk: walk file hierarchy...")
for (dirpath, dirnames, filenames) in os.walk("."):
    for dirname in dirnames:
        print("dir:", path.join(dirpath, dirname))
    for filename in filenames:
        print("file:", path.join(dirpath, filename))

print("")

print("glob.glob")
for name in glob.glob('./*.ipynb'):
    # For global paths instead, use: os.path.abspath(name)
    print(name)
```
